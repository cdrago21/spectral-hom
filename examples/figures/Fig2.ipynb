```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,10,400)

f = np.exp(-(x - 7)**2)
fd = np.exp(-(x - 5)**2)
fdd = np.exp(-(x - 1)**2)

fig, ax = plt.subplots(nrows=3, ncols=1, figsize = (5,5))
ax[0].plot(x,f, lw = 2, color = 'black')
ax[0].plot(x,f, lw = 2, color = 'red', ls = '--')

ax[1].plot(x,f, lw = 2, color = 'black')
ax[1].plot(x,fd, lw = 2, color = 'red', ls = '--')

ax[2].plot(x,f, lw = 2, color = 'black')
ax[2].plot(x,fdd, lw = 2, color = 'red', ls = '--')
for i in [0,1,2]:
    ax[i].axis('off')
    
ax[0].text(-5, 0.75,'indistinguishable', fontsize = 12)
ax[1].text(-5, 0.75,'partially distinguishable', fontsize = 12)
ax[2].text(-5, 0.75,'distinguishable', fontsize = 12)
plt.subplots_adjust(wspace=None, hspace=0.4)

# plt.annotate(s='test', xy=(10,0.5), xytext=(-4,-0.5), arrowprops=dict(arrowstyle='->'))
fig.savefig('figures/Fig2.pdf', dpi=300, bbox_inches='tight')
plt.show()

# ax.get_xaxis().set_visible(False)
# ax.get_yaxis().set_visible(False)
```
